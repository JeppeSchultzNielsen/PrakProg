```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def circ(x,y):
    if(x**2+y**2 > 0.95 and x**2+y**2 < 1): 
        return True
    else: return False

xs_circ = []
ys_circ = []
for x in np.linspace(-1,1,1000):
    for y in np.linspace(-1,1,1000):
        if(circ(x,y)):
            xs_circ.append(x)
            ys_circ.append(y)
```

```python
data = np.transpose(np.loadtxt("test.txt"))
xs = data[0]
ys = data[1]

fig, ax = plt.subplots(figsize = [10,10])

ax.scatter(xs,ys,marker = '.', color = 'black')

ax.scatter(xs_circ,ys_circ,marker = '.', color = 'red')
ax.set_xlim(0,1)
ax.set_ylim(0,1)
```
