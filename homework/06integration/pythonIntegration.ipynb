```python
import scipy.integrate as integrate
import numpy as np
```

```python
result = integrate.quad(lambda x: 1/np.sqrt(x), 0, 1, full_output=100)
result
```

```python
result = integrate.quad(lambda x: np.log(x)/np.sqrt(x), 0, 1, full_output=100)
result
```

```python
result = integrate.quad(lambda x: np.exp(-x*x), -np.inf, np.inf, full_output=100)
result
```

```python
result = integrate.quad(lambda x: np.exp(-x), -1, np.inf, full_output=100)
result
```

```python
result = integrate.quad(lambda x: np.exp(x), -np.inf, 1, full_output=100)
result
```

```python
result = integrate.quad(lambda x: np.sin(x)/x, -np.inf, np.inf, full_output=100)
result
```
